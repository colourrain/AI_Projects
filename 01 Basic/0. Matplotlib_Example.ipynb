```python
from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
#调用math.pi方法弧度转为角度
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.subplot(121)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
#使用show展示图像
plt.subplot(122, facecolor = 'b') 
plt.plot(x, x**2, color ='r')
plt.show()
```
