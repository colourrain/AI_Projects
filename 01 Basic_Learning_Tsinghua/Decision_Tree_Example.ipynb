```python
# Decision Tree
# 基于iris_data.csv数据建立决策树模型
```

```python
# load the data
import pandas as pd
import numpy as np
data = pd.read_csv("iris_data1.csv")
data.head()
```

```python
# define x, y
x = data.drop(['Species'],axis=1)
y = data.loc[:,'Species']
print(np.unique(y))
```

```python
# establish the decision tree
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1)
dc_tree.fit(x,y)
```

```python
# evaluate the accuancy
y_predict = dc_tree.predict(x)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
#visualize the decision tree
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(7,7))
tree.plot_tree(dc_tree,filled=True, feature_names=['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'], class_names=['setosa','versicolor','virginica'])
plt.show()
```
