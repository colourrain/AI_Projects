```python
from pptx import Presentation
from pptx.enum.shapes import MSO_SHAPE_TYPE
ppt = Presentation('./test.pptx')
```

```python
for slide_number, slide in enumerate(ppt.slides, start=1):
    print(f"Slide {slide_number}:")
    for shape in slide.shapes:
        if shape.has_text_frame:
            print(shape.text)

        if shape.has_table:
            table = shape.table
            for row_idx, row in enumerate(table.rows):
                for col_idx, cell in enumerate(row.cells):
                    cell_text = cell.text
                    print(f"Row {row_idx + 1}, Column {col_idx + 1}: {cell_text}")

        if shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
            image = shape.image
            image_filename = './test.jpg'
            with open(image_filename, "wb") as f:
                f.write(image.blob)
```

```python
from docx import Document
```

```python
def read_docx(file_path):
    doc = Document(file_path)
    for para in doc.paragraphs:
        print(para.text)

    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                print(cell.text, end='|')
            print()


file_path = './test.docx'
read_docx(file_path)
```

```python
# cunk textsplit
# 递归文本分块 RecursiveCharacterTextSplitter
import numpy as np
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_experimental.text_splitter import SemanticChunker
```

```python
doc_path="./山鹰国际控股股份公司祥恒包装集团原料管理制度.pdf"
loader = PyMuPDFLoader(doc_path)
docs = loader.load()
```

```python
data =[]
for doc in docs:
    data += doc.page_content.split('\u3002')
```

```python
word_num = [len(txt) for txt in data if len(txt) != 0 ]
```

```python
np.mean(word_num)
```

```python
np.min(word_num)
```

```python
np.max(word_num)
```

```python
np.percentile(word_num, q=95)
```

```python
r_spliter  = RecursiveCharacterTextSplitter(chunk_size=200,
                                         chunk_overlap=60,
                                         separators=["\n\n",
                                                    "\n",
                                                    ".",
                                                    "\uff0e",
                                                    "\u3002",
                                                    ",",
                                                    "\uff0c",
                                                    "\u3001"
                                                    ])
```

```python
split_docs = loader.load_and_split(r_spliter)
```

```python
print(split_docs[0].page_content)
```

```python
print(split_docs[1].page_content)
```

```python
docs = loader.load()
split_docs = r_spliter.split_documents(docs)
```

```python
# 基于向量语义的分块
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
```

```python
model_path = '../model/gte-large-zh/'
embeddings = HuggingFaceEmbeddings(model_name = model_path, model_kwargs={'device':'cpu'})
```

```python
semantic_chunker = SemanticChunker(embeddings,
                                  breakpoint_threshold_type='percentile',
                                  sentence_split_regex='(？=[。,；：])',
                                  buffer_size=2)
```

```python
docs[5].page_content
```

```python
semantic_docs =semantic_chunker.create_documents([docs[5].page_content])
```

```python
for semantic_chunk in semantic_docs:
    print(semantic_chunk.page_content)
    print(len(semantic_chunk.page_content))
    print('--'*50)
```
