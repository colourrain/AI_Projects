```python
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain_community.llms import Ollama
```

```python
llm = Ollama(model="qwen2.5:3b", temperature=0.1)
```

```python
from langchain.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(
    "what is the best name to describe a company that makes {product}?\n\n"
    "The name should be short, catchy, and easy to remember.\n\n")
```

```python
chain = prompt | llm
```

```python
product = "a new kind of electric car"
chain.invoke(input=product)
```
